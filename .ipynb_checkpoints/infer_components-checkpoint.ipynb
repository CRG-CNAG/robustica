```python
import os
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA, FastICA

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load data
data_file = os.path.join('data','Sastry2019.xlsx')
data = pd.ExcelFile(data_file)
data.sheet_names
```

```python
controls = [c for c in X.columns if 'control' in c]
controls
```

```python
# get data
X = data.parse('Expression Data').set_index('log-TPM')

# compute log2FCs with respect to controls
X = X - X[controls].mean(1).values.reshape(-1,1)

X
```

```python
def TSS(x):
    """Total Sum Squared"""
    return np.sum(x**2)

def calculate_explained_variance(X, S, A, k):
    """Explained variance by a component k"""
    return 1 - (TSS(X - S[:,k].reshape(-1,1).dot(A[:,k].reshape(-1,1).T)) / TSS(X))
```

```python
# how many components explain a % of variance
pca = PCA()
pca.fit(X)
inferred_components = np.min(np.where(np.cumsum(pca.explained_variance_ratio_)>0.99))
inferred_components
```

```python
%%time

n_components = [10,25,50,100,inferred_components]#,200,250,(X.shape[1])]

results = {}
for n in n_components:
    # run ICA with all components
    ica = FastICA(n_components = n, tol=1e-8, max_iter=int(1e10))
    S = ica.fit_transform(X)
    A = ica.mixing_

    # sort by decreasing explained variance
    explained_variance = np.array([calculate_explained_variance(X.values, S, A, k) for k in range(S.shape[1])])
    explained_variance_ratio = explained_variance / np.sum(explained_variance)
    idx = np.argsort(explained_variance_ratio)[::-1]

    explained_variance = explained_variance[idx]
    explained_variance_ratio = explained_variance_ratio[idx]
    S = S[:,idx]
    A = A[:,idx]
    
    result = {'S':S,'A':A,'explained_variance':explained_variance,'explained_variance_ratio':explained_variance_ratio}
    results[n] = result
```

```python
ica.n_iter_
```

## The more components the more equal is variance explained per component

```python
x = {n:results[n]['explained_variance_ratio'] for n in results.keys()}
```

```python
fig, ax = plt.subplots(2,4,figsize=(10,5))
ax = ax.ravel()

for n,i in zip(x.keys(),ax):
    g = sns.histplot(ax=i, x = x[n])
    g.set_title('n components: %s'%n)
    
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,4,figsize=(10,5))
ax = ax.ravel()

for n,i in zip(x.keys(),ax):
    g = sns.scatterplot(ax=i, x = np.arange(len(x[n])), y = np.cumsum(x[n]))
    g.set_title('total components: %s'%n)
    g.axhline(0.9, linestyle='dashed', color='black')
    
plt.tight_layout()
plt.show()
```

It seems like if there are very few components the cumulative explained variance ratio increases lineary and if there are too many every component has the same weight as we are creating a "synonymous matrix".

The number of components necessary to explain 90% of the variance depends on the number of components that we choose at the beginning.
